### Data Quality
<b>Data Quality Dimensions/Aspects/Metrics</b>
The 6 core dimensions are :
<ol>
<li>Completeness</li>
<li>Uniqueness</li>
<li>Timeliness</li>
<li>Validity</li>
<li>   Accuracy</li>
<li> Consistency</li>
</ol>
But mainly the following aspects/issues are more important and listed in decreasing order in order of severity:
<ol>
    <li>Completeness : Missing data
        <li>Validity : Invalid data, problems with data, eg. negative height. Or duplicate records Primary Keys ensure this.
            <li>Accuracy : Wrong data that is valid. Eg. wrong height for a certain weight
                <li>Consistency: Inconsistent data is valid and accurate. Eg. Full state name vs abbreviated form.
</ol>
